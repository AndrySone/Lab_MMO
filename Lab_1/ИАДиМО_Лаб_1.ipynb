```python
data = pd.read_excel("/content/drive/MyDrive/ИАДиМО/Лаб_1/Инд задание/Raisin_Dataset.xlsx")

print(data)
```

```python
# Извлечение необходимых данных
kecimen_data = data[data['Class'] == 'Kecimen']
besni_data = data[data['Class'] == 'Besni']

# Функция для создания графиков
def create_scatter_plot(x, y, x_label, y_label, title):
    plt.figure()
    plt.scatter(kecimen_data[x], kecimen_data[y], color='red', label='Kecimen', alpha=0.5)
    plt.scatter(besni_data[x], besni_data[y], color='green', label='Besni', alpha=0.5)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid()
    plt.show()

# Список всех характеристик для сравнения
features = ['Area','MajorAxisLength', 'MinorAxisLength', 'Eccentricity', 'ConvexArea', 'Extent', 'Perimeter']

# Создание всех возможных сочетаний характеристик
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        create_scatter_plot(features[i], features[j], features[i], features[j], f'{features[i]} vs {features[j]}')
```
